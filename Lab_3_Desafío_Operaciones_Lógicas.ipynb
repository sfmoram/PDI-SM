   # <font color='darkred'> Importación de  Líbrerias

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

   # <font color='darkred'> Cargar la imágen

```python
!wget https://www.dropbox.com/scl/fi/a3lwyqdj8ket92442yax8/RxTry2.png?rlkey=ffe4no01047vp8o4umwdr11ze&dl=0
!wget https://www.dropbox.com/s/716xrhlus0z33f3/Rx1.jpg
```

   # <font color='darkred'> Visualizar las imágenes

```python
X = cv2.imread('RxTry2.png?rlkey=ffe4no01047vp8o4umwdr11ze',0)
print('size = ',X.shape)
plt.figure(figsize=(10,10))
plt.imshow(X,cmap='gray')
plt.show()

Y = cv2.imread('Rx1.jpg',0)
print('size = ',Y.shape)
plt.figure(figsize=(10,10))
plt.imshow(Y,cmap='gray')
plt.show()
```

   # <font color='darkred'> Operación AND

```python
result = cv2.bitwise_and(X, Y)
plt.figure(figsize=(10,10))
plt.imshow(result, cmap='gray')
plt.show()
```

   # <font color='darkred'> Operación XOR

```python
result_xor = cv2.bitwise_xor(X, Y)
plt.figure(figsize=(10,10))
plt.imshow(result_xor, cmap='gray')
plt.show()
```

   # <font color='darkred'> Operación NOT

```python
result_not_Y = cv2.bitwise_not(Y)
plt.figure(figsize=(10,10))
plt.imshow(result_not_Y, cmap='gray')
plt.show()
```

   # <font color='darkred'> Operación OR

```python
result_or = cv2.bitwise_or(X, Y)
plt.figure(figsize=(10,10))
plt.imshow(result_or, cmap='gray')
plt.show()
```
